```python
import pandas as pd 
import matplotlib.pyplot as plt

#sample dataset 
data = {"Category": ["Food", "Rent", "Transport","Food","Rent","Transport"], "Amount":[20,800,50,30,800,60]}

df = pd.DataFrame(data)


#show data
print(df)

#Group by 
summary = df.groupby("Category")["Amount"].sum()
print(summary)

#Plot
summary.plot(kind="bar")
plt.title("Expenses by Category")
plt.show()

```
